Analyse descriptive de tes 300 153 réservations révèle un marché du billet d'avion extrêmement polarisé, où la moyenne de 20 889 est fortement distordue par des valeurs extrêmes. Le fait que la médiane (7 425) soit trois fois inférieure à la moyenne indique une asymétrie positive majeure : la grande majorité des passagers voyage à bas prix, tandis qu'une minorité (le dernier quart des données commençant à 42 521) paie des tarifs premium pouvant atteindre 123 071. Cette dispersion massive, confirmée par un écart-type très élevé de 22 697, montre que des facteurs comme la classe (Business vs Economy) et l'anticipation de l'achat (entre 1 et 49 jours) ne sont pas de simples variables, mais les véritables moteurs qui fragmentent ton dataset en deux segments de clientèle totalement distincts.

Persona :

On distingue deux profils types : d'un côté, l'Explorateur Économe, qui représente la masse des données, réserve presque deux mois à l'avance (49 jours) et voyage en classe Economy pour minimiser ses coûts. De l'autre, le Professionnel Pressé, qui appartient au dernier quart du dataset (les prix au-dessus de 42 521), réserve dans l'urgence (1 à 7 jours avant) et privilégie la classe Business peu importe le tarif.
