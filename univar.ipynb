```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('Clean_Dataset.csv')

# ── 2 listes ──
discretes = ['airline', 'source_city', 'departure_time', 'stops', 'arrival_time', 'destination_city', 'class']
continues = ['price', 'duration']  

# ── loop discrètes ──
for col in discretes:
    df[col].value_counts().sort_index().plot(kind="bar", title=col)
    plt.show()

# ── loop continues ──
for col in continues:
    fig, ax = plt.subplots(1, 2)
    sns.histplot(df[col], ax=ax[0])
    sns.boxplot(df[col], ax=ax[1])
    plt.suptitle(col)
    plt.show()
```
